```python
# ================================
# 1. SETUP
# ================================
# Install (run once)
# pip install langchain langchain-openai openai faiss-cpu tiktoken

import os
from langchain_openai import ChatOpenAI

# Set your OpenAI API Key
os.environ["OPENAI_API_KEY"] = "your_api_key_here"

# Initialize LLM
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.7
)

# ================================
# 2. BASIC LLM CALL
# ================================
print("\n=== Basic LLM Call ===")
response = llm.invoke("What is LangChain?")
print(response.content)


# ================================
# 3. PROMPT TEMPLATE
# ================================
from langchain.prompts import PromptTemplate

print("\n=== Prompt Template ===")

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

formatted_prompt = template.format(topic="Artificial Intelligence")
print(formatted_prompt)


# ================================
# 4. SIMPLE CHAIN
# ================================
from langchain.chains import LLMChain

print("\n=== LLM Chain ===")

chain = LLMChain(llm=llm, prompt=template)

chain_output = chain.run("Machine Learning")
print(chain_output)


# ================================
# 5. MEMORY (Conversation)
# ================================
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

print("\n=== Memory Example ===")

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=False
)

conversation.run("Hi, my name is Rocky")
response = conversation.run("What is my name?")
print(response)


# ================================
# 6. TOOL (Custom Function)
# ================================
print("\n=== Tool Example ===")

def calculator_tool(expression: str):
    try:
        return str(eval(expression))
    except:
        return "Error in calculation"


# ================================
# 7. AGENT WITH TOOL
# ================================
from langchain.agents import initialize_agent, Tool

print("\n=== Agent Example ===")

tools = [
    Tool(
        name="Calculator",
        func=calculator_tool,
        description="Useful for solving math expressions"
    )
]

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=False
)

agent_response = agent.run("What is 45 * 12?")
print(agent_response)


# ================================
# 8. DOCUMENT LOADER + VECTOR STORE (OPTIONAL BONUS)
# ================================
from langchain.document_loaders import TextLoader
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings

print("\n=== Document + Vector Store ===")

# Create a sample text file before running
# Example: data.txt

try:
    loader = TextLoader("data.txt")
    documents = loader.load()

    vectorstore = FAISS.from_documents(
        documents,
        OpenAIEmbeddings()
    )

    print("Vector store created successfully!")

except Exception as e:
    print("Skipping vector store (no file found)")


# ================================
# END
# ================================
```
