1. Introduction to LangChain

What is LangChain?

LangChain is an open-source framework designed to simplify the development of applications powered by Large Language Models (LLMs). Instead of calling an LLM directly, LangChain helps developers build structured, modular pipelines that combine LLMs with external tools, memory, and data sources.

Why is LangChain Important?
Modern LLMs like GPT are powerful, but raw API calls are not enough for real-world applications. Problems arise such as:

No memory of past interactions
Difficulty integrating APIs/tools
Lack of structured workflows
Repetitive prompt engineering
LangChain solves these by introducing abstraction layers.

Problems It Solves
✅ Orchestration of multiple LLM calls
✅ Chaining tasks together
✅ Integration with tools (APIs, databases)
✅ Stateful conversations (memory)

2. Core Components of LangChain

Concept

LLMs are the core engines that generate responses. LangChain supports models from OpenAI, HuggingFace, etc.

Why it Exists

To standardize how different models are used.

Code Example

```python
!pip install langchain-openai
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",   # better than gpt-3.5-turbo (now deprecated in many cases)
    temperature=0.7
)

response = llm.invoke("Explain LangChain in simple terms")
print(response.content)
```

2.2 Prompts and Prompt Templates
Concept

Prompt templates allow dynamic generation of prompts using variables.

Why it Exists

Hardcoding prompts is inefficient. Templates make prompts reusable.

Code Example

```python
from langchain.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

prompt = template.format(topic="LangChain")
print(prompt)
```

2.3 Chains
Concept

Chains combine multiple components into a pipeline.

Why it Exists

To structure workflows instead of making isolated API calls.

Code Example

```python
from langchain.chains import LLMChain

chain = LLMChain(llm=llm, prompt=template)

output = chain.run("Artificial Intelligence")
print(output)
```

2.4 Memory
Concept

Memory stores previous interactions for context-aware responses.

Why it Exists

LLMs are stateless by default.

Code Example

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory
)

conversation.run("Hi, I am Rocky")
conversation.run("What is my name?")
```

2.5 Agents
Concept

Agents decide what action to take dynamically.

Why it Exists

Instead of fixed chains, agents enable reasoning and decision-making.

Code Example

```python
from langchain.agents import initialize_agent, Tool

def calculator_tool(input):
    return str(eval(input))

tools = [
    Tool(
        name="Calculator",
        func=calculator_tool,
        description="Useful for math calculations"
    )
]

agent = initialize_agent(tools, llm, agent="zero-shot-react-description")

agent.run("What is 25 * 4?")
```
